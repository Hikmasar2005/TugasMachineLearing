```python
import pandas as pd

#baca file csv
df = pd.read_csv('/content/cars.csv')
df.head(3)
df.info()
```

```python
df.head()
```

```python
X = df.drop([' cubicinches', ' weightlbs', ' brand'], axis=1)
X.head()
```

```python
from sklearn.cluster import KMeans

#membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)
```

```python
X.head()
```

```python
# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_
```

```python
X.head()
```

```python
# Membuat plot KMeans dengan 5 klaster untuk atribut mpg dan cylinders
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X['mpg'], y=X[' cylinders'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('mpg')
plt.ylabel('cylinders')
plt.show()
```

```python
# Membuat plot KMeans dengan 5 klaster untuk atribut mpg dan hp
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X['mpg'], y=X[' hp'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('mpg')
plt.ylabel('hp')
plt.show()
```

```python
# Membuat plot KMeans dengan 5 klaster untuk atribut cylinders dan time-to-60
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X[' cylinders'], y=X[' time-to-60'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('cylinders')
plt.ylabel('time-to-60')
plt.show()
```

```python
# Membuat plot KMeans dengan 5 klaster untuk atribut hp dan year
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X[' hp'], y=X[' year'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('hp')
plt.ylabel('year')
plt.show()
```
